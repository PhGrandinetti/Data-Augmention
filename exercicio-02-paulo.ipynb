```python
import gradio as gr
import cv2
import numpy as np

#Função de transformação das imagens
def transformar(img, angulo, escala, tx, ty, gamma, flip):

    if img is None:
        return None, None

    img_aug = img.copy()

    h, w = img_aug.shape[:2]

    M_rot = cv2.getRotationMatrix2D((w/2, h/2), angulo, escala)
    img_aug = cv2.warpAffine(img_aug, M_rot, (w, h))

    M_trans = np.float32([[1, 0, tx], [0, 1, ty]])
    img_aug = cv2.warpAffine(img_aug, M_trans, (w, h))

    if flip:
        img_aug = cv2.flip(img_aug, 1)

    img_aug = np.power(img_aug/255.0, gamma)
    img_aug = (img_aug * 255).astype(np.uint8)

    return img, img_aug #Retornando a original e a trasnformada

#Gradio
interface = gr.Interface(
    fn=transformar,
    inputs=[
        gr.Image(type="numpy"),
        gr.Slider(-180, 180, value=0, label="Rotação"),
        gr.Slider(0.5, 2.0, value=1.0, label="Escala"),
        gr.Slider(-100, 100, value=0, label="Translação X"),
        gr.Slider(-100, 100, value=0, label="Translação Y"),
        gr.Slider(0.3, 3.0, value=1.0, label="Gamma"),
        gr.Checkbox(label="Flip Horizontal")
    ],
    outputs=[
    gr.Image(type="numpy", label="Original"),
    gr.Image(type="numpy", label="Transformada")
    ], title="Visualização de Data Augmentation",
    description="Ajuste os parâmetros para aplicar transformações geométricas e de iluminação na imagem em tempo real."
)

interface.launch()
```
